```python
def math(x: float, y: float, func: str) -> float:
    """Function does a calculation based on the given function"""

    if func == 'add':
        return x + y
    elif func == 'distract':
        return x - y
    elif func == 'multiply':
        return x * y
    elif func == 'divide':
        if y != 0:
            return x / y
        else:
            return 0
    elif func == "modulo":
        return x % y
    else:
        return 0
```

```python
var_math = math(param_var_x, param_var_y, param_func)
print(var_math)
```
